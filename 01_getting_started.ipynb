## Generating own dataset

#### 1) Linearly distributed datasets

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_regression
```

```python
X, Y = make_regression(n_samples=100, n_features=1, noise=5.5)
X.shape, Y.shape
```

```python
plt.scatter(X, Y)
```

```python
X, Y = make_regression(n_samples=100, n_features=3, noise=5.5)
X.shape, Y.shape
```

```python
plt.scatter(X[:, 1], Y)
```

#### 2.1) Clustered dataset

```python
from sklearn.datasets import make_blobs
X, Y = make_blobs(500, centers=5) # generates isotropic gaussian blobs for clustering
X.shape, Y.shape
```

```python
rgb = np.array(['r', 'g', 'b', 'y', 'm'])
plt.scatter(X[:,0], X[:,1], color=rgb[Y])
```

#### 2.2) Clustered dataset distributed in circular fassion

```python
from sklearn.datasets import make_circles
X, Y = make_circles(n_samples=500, noise=0.15) # generates isotropic gaussian blobs for clustering
X.shape, Y.shape
```

```python
rgb = np.array(['r', 'g', 'b', 'y', 'm'])
plt.scatter(X[:,0], X[:,1], color=rgb[Y])
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
heights = np.array([[1.6], [1.65], [1.7], [1.73], [1.8]])
weights = np.array([[60], [65], [72.3], [75], [80]])
heights.shape, weights.shape
```

```python
plt.scatter(heights, weights)
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.grid(True)
```

```python
model = LinearRegression()
model.fit(heights, weights)
```

```python
pred = model.predict(np.array([[1.75]]))
pred
```

```python
plt.scatter(heights, weights)
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.plot(heights, model.predict(heights))
plt.grid(True)
plt.axis([0, 2, -200, 200])
```

```python
# Getting the gradient and intercept of the linear regression line
plt.scatter(heights, weights)
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.plot(heights, model.predict(heights))
plt.grid(True)
plt.axis([0, 2, -200, 200])

extreme_heights = np.array([[0], [2]])
plt.plot(extreme_heights, model.predict(extreme_heights), 'r')
```

```python
# y-intercept
round(model.predict(extreme_heights)[0,0], 4)
```

```python
print(model.intercept_[0]) # y-intercept
print(model.coef_[0])  # to get the coefficients
```
